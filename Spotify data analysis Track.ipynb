```python
dataset_track.set_index('release_date',inplace=True)
dataset_track.index = pd.to_datetime(dataset_track.index)
dataset_track.head()
```

```python
dataset_track.iloc[18]
```

```python
dataset_track["duration"]=dataset_track['duration_ms'].apply(lambda x:round(x/1000))
dataset_track.drop('duration_ms',inplace=True,axis=1)
```

```python
dataset_track.head()
```

```python
corr_df = dataset_track.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
```

```python
type(corr_df)
```

```python
corr_df
```

```python
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot=True,fmt='0.1g',linecolor='Black',linewidths=1)
heatmap.set_title('Correlation Heatmap BTW Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = dataset_track.sample(int(0.004*len(dataset_track)))
```

```python
sample_df.shape
```

```python
sample_df.head()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x='energy', y='loudness', color='black').set_title('Loudnes v/s Energy')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x='acousticness', y='popularity', color='black').set_title('Acousticness v/s Popularity')
```

```python
dataset_track['dates'] = dataset_track.index.get_level_values('release_date')
```

```python
dataset_track['dates'] = pd.to_datetime(dataset_track['dates'])
```

```python
dataset_track['dates'].head()
```

```python
years = dataset_track['dates'].dt.year
months = dataset_track['dates'].dt.month
days = dataset_track['dates'].dt.day
```

```python
print(years,months,days)
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title="Number of songs Per Year")
```

```python
total_dr = dataset_track['duration']
```

```python
plt.figure(figsize=(24,12))
Plot = sns.barplot(x=years,y=total_dr)
Plot.set(title='Duration Vs Year',xlabel='Years')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set_style(style='whitegrid')
Plot = sns.lineplot(x=years,y=total_dr)
Plot.set(title='Duration Vs Year',xlabel='Years')
plt.show()
```
